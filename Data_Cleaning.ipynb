```python
data_frame['text'] = data_frame['text'].apply(lambda x : ' '.join([word for word in str(x).split() if not word in set(stopwords("el"))]))
```

```python
from nltk.stem import WordNetLemmatizer 

#Lemmitization
lemmatizer = WordNetLemmatizer()
data_frame['text'] = data_frame['text'].apply(lambda x : ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
```

```python
from nltk.stem import PorterStemmer
#Stemming
ps = PorterStemmer()
adwait = data_frame
#adwait.head()
data_frame['text'] = data_frame['text'].apply(lambda x : ' '.join([ps.stem(word) for word in x.split()]))
```

```python
data_frame.head()
```

```python
#Tokenization
corpus = []
for i in range(0,1776):
    tweet = data_frame['text'][i]
    tweet = tweet.lower()
    tweet = tweet.split()
    tweet = [ps.stem(word) for word in tweet if not word in set(stopwords("el"))]
    tweet = ' '.join(tweet)
    corpus.append(tweet)
```

```python
#Ensuring all the tweets are tokenized into individual words
len(corpus)
```
